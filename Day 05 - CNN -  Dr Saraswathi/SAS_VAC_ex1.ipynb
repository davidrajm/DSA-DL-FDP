```python
#option 2

import numpy as np
import pandas as pd

# Step 1: Dataset
X = np.array([
    [45, 0.80, 9120],
    [40, 0.12, 2000],
    [38, 0.08, 3042],
    [25, 0.03, 3300],
    [49, 0.02, 63588],
    [74, 0.37, 3500]
])

# Actual values (Column W)
y = np.array([1, 1, 0, 0, 0, 0])

# Step 2: Normalize
X_norm = X / np.max(X, axis=0)

# Step 3: Manual weights (to get mixed predictions)
weights = np.array([0.2, 3.0, -2.5])
bias = -0.5

# Step 4: Sigmoid
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Step 5: Predictions
predictions = []
errors = []

for i in range(len(X_norm)):
    z = np.dot(X_norm[i], weights) + bias
    pred = sigmoid(z)
    predictions.append(round(pred, 2))
    errors.append(round(y[i] - pred, 2))  # same as your table

# Step 6: Create DataFrame
df = pd.DataFrame({
    "Customer ID": range(1, 7),
    "Customer Age": X[:, 0],
    "Debt Ratio": X[:, 1],
    "Monthly Income": X[:, 2],
    "Loan Defaulter (W)": y,
    "Default Prediction (X)": predictions,
    "Prediction Error": errors
})

# Step 7: Print nicely
print(df)
```

```python
def highlight(val):
    if val > 0:
        return 'color: green'
    else:
        return 'color: red'

df.style.applymap(highlight, subset=["Prediction Error"])
```

```python
# MLP

import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Step 1: Dataset
X = np.array([
    [45, 0.80, 9120],
    [40, 0.12, 2000],
    [38, 0.08, 3042],
    [25, 0.03, 3300],
    [49, 0.02, 63588],
    [74, 0.37, 3500]
])

y = np.array([1, 1, 0, 0, 0, 0])

# Step 2: Normalize
X = X / np.max(X, axis=0)

# -------------------------------
# 🔷 MODEL 1: SINGLE LAYER
# -------------------------------
model1 = Sequential([
    Dense(1, input_dim=3, activation='sigmoid')
])

model1.compile(optimizer='adam', loss='binary_crossentropy')
model1.fit(X, y, epochs=100, verbose=0)

pred1 = model1.predict(X).flatten()

# -------------------------------
# 🔷 MODEL 2: MLP
# -------------------------------
model2 = Sequential([
    Dense(4, input_dim=3, activation='relu'),  # hidden layer
    Dense(1, activation='sigmoid')
])

model2.compile(optimizer='adam', loss='binary_crossentropy')
model2.fit(X, y, epochs=100, verbose=0)

pred2 = model2.predict(X).flatten()

# -------------------------------
# 🔷 CREATE COMPARISON TABLE
# -------------------------------
df = pd.DataFrame({
    "Customer ID": range(1, 7),
    "Actual": y,
    "Single Layer Output": np.round(pred1, 2),
    "MLP Output": np.round(pred2, 2)
})

print(df)
```

```python
df["Single Layer Class"] = df["Single Layer Output"].apply(lambda x: 1 if x > 0.5 else 0)
df["MLP Class"] = df["MLP Output"].apply(lambda x: 1 if x > 0.5 else 0)

print(df)
```
